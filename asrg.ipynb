```python
# Cell 3: Model Initialization

import torch
from transformers import pipeline

# Use a smaller model like 'small' for speed in Colab, or 'base'/'medium' for better accuracy
# 'whisper-medium' is often a good balance for multilingual tasks.
MODEL_NAME = "openai/whisper-medium" # Strong multilingual model

# Check for GPU and set device
device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# Initialize the ASR pipeline
# Task: 'automatic-speech-recognition'
# Model: Whisper-medium
# Device: GPU (if available)
asr_pipeline = pipeline(
    "automatic-speech-recognition",
    model=MODEL_NAME,
    chunk_length_s=30, # Recommended for long-form audio
    device=device
)

print(f"\nASR Pipeline initialized with model: {MODEL_NAME}")
```

```python
# Cell 4: Core Inference Function

def transcribe_and_translate_tanglish(audio_file_path):
    """
    Takes an audio file (Tanglish speech) and returns the English transcription.

    Args:
        audio_file_path (str): The file path to the uploaded audio.

    Returns:
        str: The English transcription.
    """
    if not audio_file_path:
        return "Please upload an audio file."

    # --- Step 1: Speech-to-Text and Translation (Directly using Whisper's translate task) ---
    # Whisper can be configured to transcribe (speech-to-source-language-text)
    # or translate (speech-to-English-text). For this project, we want the latter.
    try:
        # Set the language to Tamil ('ta') and the task to 'translate'
        # This tells Whisper: "This is a Tamil-based audio, please translate the content to English."
        # This is a good proxy for Tanglish -> English.
        result = asr_pipeline(
            audio_file_path,
            generate_kwargs={"language": "tamil", "task": "translate"},
            # Use a slightly faster batch size if running on CPU
            batch_size=4 if device == "cpu" else 16
        )

        english_transcription = result['text'].strip()

        return english_transcription

    except Exception as e:
        return f"An error occurred during transcription/translation: {e}"

# Test the function with a simulated file (Optional)
# if not sample_df.empty:
#     test_path = sample_df.iloc[0]['audio_path']
#     print(f"\n--- Testing with sample audio: {test_path} ---")
#     transcribed_text = transcribe_and_translate_tanglish(test_path)
#     print(f"Expected Tanglish: {sample_df.iloc[0]['tanglish_text']}")
#     print(f"Expected English: {sample_df.iloc[0]['english_transcription']}")
#     print(f"Model Output: {transcribed_text}")
# else:
#     print("\nNo sample audio generated to test.")
```
